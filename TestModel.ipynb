```python
import face_recognition
import os
from IPython.display import display, Image
```

```python
train_image_path = "/home/hany_jr/Ai/Data sets/FaceRecog/presidents/known_people"
train_data = os.listdir(train_image_path)

test_image_path = "/home/hany_jr/Ai/Data sets/FaceRecog/presidents/unknown_people"
test_data = os.listdir(test_image_path)
```

```python
# show the images
for img in train_data:
    display(
        Image(
            filename="/home/hany_jr/Ai/Data sets/FaceRecog/presidents/known_people/"
            + img
        )
    )
```

```python
for img in test_data:
    display(
        Image(
            filename="/home/hany_jr/Ai/Data sets/FaceRecog/presidents/unknown_people/"
            + img
        )
    )
```

## [predict by python]()

```python
for img in test_data:
    test_image = face_recognition.load_image_file(test_image_path + "/" + img)
    test_face_encoding = face_recognition.face_encodings(test_image)[0]

    for img2 in train_data:
        train_image = face_recognition.load_image_file(train_image_path + "/" + img2)
        train_face_encoding = face_recognition.face_encodings(train_image)[0]

        results = face_recognition.compare_faces(
            [test_face_encoding], train_face_encoding
        )
        if results[0]:
            display(
                Image(
                    filename="/home/hany_jr/Ai/Data sets/FaceRecog/presidents/known_people/"
                    + img2
                )
            )
            display(
                Image(
                    filename="/home/hany_jr/Ai/Data sets/FaceRecog/presidents/unknown_people/"
                    + img
                )
            )

            print("Match found: " + img + " is " + img2)
            break
    else:
        print("No match found for " + img)
```
